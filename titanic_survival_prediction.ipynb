```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import MultinomialNB, ComplementNB, GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
train_df = pd.read_csv("./train.csv")
test_df = pd.read_csv("./test.csv")

train_df.head()
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())

train_df.isna().sum()
```

```python
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())
test_df.isna().sum()

test1 = test_df[["Pclass", "Age", "SibSp", "Parch"]].values

test1
```

```python
# converting the training data into inputs and outputs

x = train_df[["Pclass", "Age", "SibSp", "Parch"]].values
y = train_df[["Survived"]].values

x
```

```python
# splitting the training data to run the model for accuracy checking

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)

y_train
```

```python
# training the machine using mulitple models and checking the accuracy scores to decide which model suits best

models = [MultinomialNB(), ComplementNB(), GaussianNB(), KNeighborsClassifier(n_neighbors = 5), SVC()]

for i in models:
    model = i
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score = accuracy_score(y_test, y_pred)
    print("%.3f" % score, end="\n")
```

```python
# Decided to take Gaussian Naive Bayes model based on the above result to run the test data.

model = models[2]

model.fit(x, y)

pred = model.predict(test1)

pred

prediction = pd.DataFrame({"PassengerId": test_df["PassengerId"], "Survived": pred})

prediction
```
